## 激活函数

### Sigmoid 函数
通常在输出层，将输出映射到 (0, 1) 区间，用于二分类问题。

```python
import torch
import matplotlib.pyplot as plt
```

```python
_, axes = plt.subplots(1,2)
x = torch.linspace(-20, 20, 1000)
y = torch.sigmoid(x)
axes[0].plot(x, y)
axes[0].set_title('Sigmoid')
axes[0].grid()

x = torch.linspace(-20, 20, 1000, requires_grad=True)
torch.sigmoid(x).sum().backward()
axes[1].plot(x.detach(), x.grad)
axes[1].set_title('Sigmoid Derivative')
axes[1].grid()
```

## Tanh 函数

```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-20, 20, 1000)
y = torch.tanh(x)
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x = torch.linspace(-20, 20, 1000, requires_grad=True)
torch.tanh(x).sum().backward()
plt.plot(x.detach(), x.grad)
plt.grid()
plt.show()
```

## ReLU 函数

## Softmax 函数
通常在输出层，将输出映射到 (0, 1) 区间，用于多分类问题。\
将多分类的结果以 概率 形式输出\


```python
import torch
scores = torch.tensor([0.2, 0.02, 0.15, 0.015, 1.3, 0.5, 3.75, 0.75, 1.1])
torch.softmax(scores, dim=0)
```
